```python
from PIL import Image
```

```python
def reveal_hidden_text(image_path):
    img = Image.open(image_path)
    width, height = img.size
    binary_data = ""

    for y in range(height):
        for x in range(width):
            pixel = img.getpixel((x, y))
            for channel in range(3):  # R, G, B channels
                binary_data += str(pixel[channel] & 1)  # Extract LSB

    # Convert binary data to ASCII characters
    message = ""
    for i in range(0, len(binary_data), 8):
        byte = binary_data[i:i + 8]
        char_code = int(byte, 2)
        message += chr(char_code)

    return message
```

```python
image_path = "Pwn2Hide.jpg"
```

```python
hidden_text = reveal_hidden_text(image_path)
print("Revealed hidden text:", hidden_text)
```
